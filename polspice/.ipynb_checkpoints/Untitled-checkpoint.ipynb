```python
import camb
from ispice import ispice
import numpy as np
import healpy as hp
from astropy.io import fits
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def de_camb(dls):
    
    num_dls = len(dls)
    
    ell_max = dls[0].size+2
    ell = np.arange(ell_max-2)
    
    d_to_c = 2.*np.pi/(ell*(ell+1.))
    
    output = np.zeros(num_dls*ell.size).reshape(num_dls,ell_max-2)
    
    for i,(dl,d) in enumerate(zip(dls,d_to_c)):
        output[i] = d_to_c*dl
    return output                    
        
```

```python
pars = camb.CAMBparams()
pars.set_cosmology(H0=67., ombh2=0.02, omch2=0.12, mnu=0.06, omk=0, tau=0.06)
pars.InitPower.set_params(ns=0.965, r=0)
pars.set_for_lmax(1500, lens_potential_accuracy=0);
pars.DoLensing = True
results = camb.get_results(pars)
powers =results.get_cmb_power_spectra(pars)
```

```python
dl = powers['lensed_scalar'].swapaxes(0,1) #TT EE BB TE (CAMB order)
cl = de_camb(dl)
ell = np.arange(cl[1,:].size)
```

```python
fig, axes = plt.subplots(2,2,figsize=(10,10))
for i in range(4):
    ax = np.ravel(axes)[i]
    ax.plot(ell,dl[i,:],label='%d'%i)
    ax.legend()
```

```python
cmb = hp.synfast([dl[0,:],dl[1,:],dl[2,:],dl[3,:],
                  dl[3,:]*1e-2,dl[3,:]*1e-3],
                  256,new=True,verbose=False) #TT, EE, BB, TE, and make up EB, TB

hp.write_map('map/example.fits', cmb)

cmb = hp.read_map('map/example.fits',field=(0,1,2),verbose=False)

hp.mollview(cmb[0]);hp.mollview(cmb[1])
```

```python
ispice(mapin1="map/example.fits",
           clout="cls/example_cl.fits",
           polarization="YES",
           subav="YES",
           subdipole="YES",
           pixelfile="YES",
           )
```
